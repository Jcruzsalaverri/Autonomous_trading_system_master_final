```python

#TIME BARS

Time_bars = data1.groupby(pd.Grouper(key="timestamp", freq="15min")).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first' })


Time_bars.columns = ['open','high','low','close','size','timestamp']


Time_bars
```

#Absolute price change study (CHAPTER 3. DATA ACQUISITON)

```python
df = Time_bars
# create the bins
from plotly.subplots import make_subplots



c = Time_bars['close']
o = Time_bars['open']
diff = abs(((c / o) * 100) - 100)


Time_bars['price_abs_diff'] = diff



counts, bins = np.histogram(df.price_abs_diff, bins=np.linspace(0,2,20))
bins = 0.5 * (bins[:-1] + bins[1:])

#Time_bars['Prop_tot_price_change'] = 
#bins = np.insert(bins,0,0)
np.cumsum(counts)

#CALCULATING PROPORTION OF CANDLES AT THE RIGHT

proportion1 = (counts/5268)

proportion1 = np.insert(proportion1,0,1)

for idx, val in enumerate(proportion1):
  if idx == 0:
    proportion1[idx] = proportion1[idx]
  if idx != 0:
    proportion1[idx] = proportion1[idx-1] - proportion1[idx]

proportion1 = np.round(proportion1, 3)
prop1 = proportion1*100;
prop1 = np.round(proportion1, 3)


proportion2 = []

for idx, val in enumerate(bins):
    proportion2.append((counts[idx]*bins[idx])/1275.89473684)

np.cumsum(proportion2)

proportion2 = np.insert(proportion2,0,1)

for idx, val in enumerate(proportion2):
  if idx == 0:
    proportion2[idx] = proportion2[idx]
  if idx != 0:
    proportion2[idx] = proportion2[idx-1] - proportion2[idx]

proportion2 = np.round(proportion2, 3)
prop2 = proportion2*100;
prop2 = np.round(proportion2, 3)

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

#fig = px.bar(x=bins, y=counts, labels={'x':'Absolute price change %"', 'y':'Occurences'})
fig.add_bar(x=bins, y=counts, name="<b>Absolute price change (%)</b>")

prop12 = ['100%', '69.3%' , '42%' , '24.5%' , '14.7%' , '0.93%' , '0.61%' , '0.4%', '0.27%' ,'0.2%' , '0.17%', '0,11%', '','' , '', '','' ,'' , '','' ]
fig.add_trace(go.Scatter(
    x=bins,
    y=proportion1,
    name="Proportion of candles at the right",text=prop12,
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)
prop21 = ['', '93%' , '75.5%' , '56.5%' , '41.7%' , '31%%' , '23.4%' , '17.6%', '13.3%' ,'10.7%' , '9.2%', '6.8%', '5.4%','4.2%' , '3.4%', '2.7%','' ,'' , '','' ]

fig.add_trace(go.Scatter(
    x=bins,
    y=proportion2,
    name="Proportion price change produced by candles at the right",text=prop21,
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)


fig.update_layout(
    xaxis_title="Absolute price change (percentage)",
    yaxis_title="Frequency",
)

fig.update_layout(
    legend=dict(
        x=0.7,
        y=0.9,
        traceorder="normal",
        font=dict(
            family="sans-serif",
            size=12,
            color="black"
        ),
    )
)
fig.update_traces(textfont_size=17)


##NEED TO ADD CUSTOM TEXT TRACES TO PREVENT OVERLAPPING
##E.G.
#TEXT = ['100%', '69.3%' , '42%' , '24.5%' , '14.7%' , '0.93%' , '0.61%' , '0.4%', '0.27%' ,'0.2%' , '0.17%', '0,11%', '','' , '', '','' ,'' , '','' ]
```
