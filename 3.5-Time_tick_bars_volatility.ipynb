```python
#SNIPPET 3.1 DAILY VOLATILITY ESTIMATES

def get_daily_vol(close, lookback=100):
    """
    :param close: (data frame) Closing prices
    :param lookback: (int) lookback period to compute volatility
    :return: (series) of daily volatility value
    """
    print('Calculating daily volatility for dynamic thresholds')
    
    df0 = close.index.searchsorted(close.index - pd.Timedelta(days=1))
    df0 = df0[df0 > 0]
    df0 = (pd.Series(close.index[df0 - 1], index=close.index[close.shape[0] - df0.shape[0]:]))
        
    df0 = close.loc[df0.index] / close.loc[df0.values].values - 1  # daily returns
    df0 = df0.ewm(span=lookback).std()
    return df0
```

#Time and tick bars volatility calcs and plots

```python
Time_bars_30m = data1.groupby(pd.Grouper(key="timestamp", freq="30min")).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first' })
Time_bars_30m.columns = ['open','high','low','close','size','timestamp']

Time_bars_1h = data1.groupby(pd.Grouper(key="timestamp", freq="1h")).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first' })
Time_bars_1h.columns = ['open','high','low','close','size','timestamp']

Time_bars_4h = data1.groupby(pd.Grouper(key="timestamp", freq="4h")).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first' })
Time_bars_4h.columns = ['open','high','low','close','size','timestamp']
```

```python
# determining daily volatility using the last 50 days
Time_bars_daily_vol_30h = get_daily_vol(close=Time_bars_30m['close'], lookback=50)

Time_bars_daily_vol_1h = get_daily_vol(close=Time_bars_1h['close'], lookback=50)

Time_bars_daily_vol_4h = get_daily_vol(close=Time_bars_4h['close'], lookback=50)
```

```python

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])


fig.add_trace(go.Scatter(
    x=Time_bars_daily_vol_30h.index,
    y=Time_bars_daily_vol_30h,
    name="30min Time bars daily volatility (ADAUSDT)",
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)

fig.add_trace(go.Scatter(
    x=Time_bars_daily_vol_1h.index,
    y=Time_bars_daily_vol_1h,
    name="1H Time bars daily volatility (ADAUSDT)",
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)

fig.add_trace(go.Scatter(
    x=Time_bars_daily_vol_4h.index,
    y=Time_bars_daily_vol_4h,
    name="4H Time bars daily volatility (ADAUSDT)",
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)
```

```python
data1['count'] = 1
Time_bars_1d = data1.groupby(pd.Grouper(key="timestamp", freq="1d")).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first', 'count':'sum' })
Time_bars_1d.columns = ['open','high','low','close','size','timestamp','Daily_ticks']

avg_dly_tick = np.round(np.mean(Time_bars_1d['Daily_ticks']), decimals = -1)

avg_30m_tick = np.round(avg_dly_tick/48, decimals = -1)

avg_1h_tick = np.round(avg_dly_tick/24,decimals=-2)

avg_4h_tick = np.round(avg_dly_tick/5,decimals=-2)
```

```python
Tick_50 = data1.groupby(bar(np.arange(len(data1)), avg_30m_tick )).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first'})
Tick_50.columns = ['open','high','low','close','size','timestamp']
Tick_50.index = Tick_50['timestamp']

Tick_100 = data1.groupby(bar(np.arange(len(data1)), avg_1h_tick )).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first'})
Tick_100.columns = ['open','high','low','close','size','timestamp']
Tick_100.index = Tick_100['timestamp']

Tick_200 = data1.groupby(bar(np.arange(len(data1)), avg_4h_tick )).agg({'price': 'ohlc', 'size': 'sum', 'timestamp': 'first'})
Tick_200.columns = ['open','high','low','close','size','timestamp']
Tick_200.index = Tick_200['timestamp']
```

```python
# determining daily volatility using the last 50 days
Tick_50['vol'] = get_daily_vol(close=Tick_50['close'], lookback=50)

Tick_100['vol'] = get_daily_vol(close=Tick_100['close'], lookback=50)

Tick_200['vol'] = get_daily_vol(close=Tick_200['close'], lookback=50)
```

```python
# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])


fig.add_trace(go.Scatter(
    x=Tick_50.index,
    y=Tick_50['vol'],
    name="30min Tick bars daily volatility (ETHUSD)",
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)

fig.add_trace(go.Scatter(
    x=Tick_100.index,
    y=Tick_100['vol'],
    name="1H Tick bars daily volatility (ETHUSD)",
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)

fig.add_trace(go.Scatter(
    x=Tick_200.index,
    y=Tick_200['vol'],
    name="4H Tick bars daily volatility (ETHUSD)",
    mode = 'lines+text',
    textfont_family="Arial_Black"),
    secondary_y=True,

)
```
